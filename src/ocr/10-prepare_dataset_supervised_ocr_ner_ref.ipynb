````python
import json
from text_utils import (
    simplify_unicode_charset, 
    replace_annotation_codes, 
    fix_manual_ner_xml, 
    check_alignment_charset, 
    remove_xml_tags_and_entities)
````

## Gather and normalize all NER and OCR references
We remove empty tags.
We drop inconsistent entries (OCR text and NER XML).

````python
dataset = []
all_raw_ocr_txt = []
all_new_ocr_txt = []
limit = 10000000
count = 0
lost_entries = 0
bad_boxes = 0
for item_id, book, page, in_dataset in zip(
        annot_metadata["#"], annot_metadata["PDF"], annot_metadata["VUE"], annot_metadata["Inclus dataset"]):
    print("="*30)
    print(item_id, book, page, in_dataset)
    if not in_dataset:
        print("\t >> SKIPPING (not in dataset)")
        continue
    content = None
    with open(os.path.join(json_dir, f"{item_id:04d}.json")) as infile:
        content = json.load(infile)
    for elem in content:
        if count > limit:
            break
        if elem["type"] != "ENTRY":
            continue
        if "checked" not in elem or not elem["checked"]:
            continue
        count += 1

        newelem = {}
        for k in ["id", "box"]:
            newelem[k] = elem[k]
        newelem["book"] = book
        newelem["page"] = page

        entry_id = elem["id"]

        x, y, w, h = elem["box"]
        valid_box = x >= 0 and y >= 0 and w > 10 and h > 10
        newelem["valid_box"] = valid_box
        if not valid_box:
            bad_boxes += 1
            print(f"ERROR: ({entry_id}) bad box: {x:.0f}, {y:.0f}, {w:.0f}, {h:.0f}")
            # We keep these entries for now, as NER can be valid
            # We tag them for later filtering (for OCR eval)

        orig_ocr_txt = elem["text_ocr"].strip()
        new_ocr_txt = simplify_unicode_charset(replace_annotation_codes(orig_ocr_txt))
        newelem["text_ocr"] = new_ocr_txt
        if not check_alignment_charset(new_ocr_txt):
            print(f"FATAL: ({entry_id}) Bad charset")
            raise RuntimeError("Should not have bad charset after normalization.")
        
        orig_ner_xml = elem["ner_xml"].strip()
        new_ner_xml = fix_manual_ner_xml(orig_ner_xml)
        newelem["ner_xml"] = new_ner_xml

        orig_ner_text = remove_xml_tags_and_entities(orig_ner_xml)
        if not orig_ner_text == orig_ocr_txt:
            print(f"ERROR: ({entry_id}) NER and OCR don't match.")
            print(".....: OCR: \t%s"%(orig_ocr_txt.replace('\n', '\u2424')))
            print(".....: NER: \t%s"%(orig_ner_text.replace('\n', '\u2424')))
            lost_entries += 1
            continue

        dataset.append(newelem)
        all_raw_ocr_txt.append(orig_ocr_txt)
        all_new_ocr_txt.append(new_ocr_txt)
        
````
